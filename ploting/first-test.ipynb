```python
#Library importation
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Data calling
datac = pd.read_csv('/home/labmwl/bug/Github/spectrometer-mwl/Spectrogram Data/just-water-16x.csv', sep=',', header=0)
data1r = pd.read_csv('/home/labmwl/bug/Github/spectrometer-mwl/Spectrogram Data/1drop-r-16x-100ml.csv', sep=',', header=0)
data1b = pd.read_csv('/home/labmwl/bug/Github/spectrometer-mwl/Spectrogram Data/1drop-b-16x-100ml.csv', sep=',', header=0)
data10_100r = pd.read_csv('/home/labmwl/bug/Github/spectrometer-mwl/Spectrogram Data/10-100-r-16x-100ml.csv', sep=',', header=0)
data10_100b = pd.read_csv('/home/labmwl/bug/Github/spectrometer-mwl/Spectrogram Data/10-100-b-16x-100ml.csv', sep=',', header=0)
data1_100r = pd.read_csv('/home/labmwl/bug/Github/spectrometer-mwl/Spectrogram Data/1-100-r-16x-100ml.csv', sep=',', header=0)
data1_100b = pd.read_csv('/home/labmwl/bug/Github/spectrometer-mwl/Spectrogram Data/1-100-b-16x-100ml.csv', sep=',', header=0)
```

```python
datac
```

```python
# Clear Water
labels = datac.columns.to_list()
labels = labels[1:]
clear_bars = []
col = ['purple','blue','cyan','lime','lawngreen','gold','darkorange','red','gray','crimson']
x = np.arange(len(labels))

for i in labels:
    clear_bars.append(np.mean(datac[i].to_numpy()))

fig, ax = plt.subplots(figsize=(10,6)) 

ax.bar(labels, clear_bars, align='center', color=col)
ax.set_xticks(x, labels, rotation=90)
plt.ylabel('Intensity')
plt.title('Clear Water')
for i in ax.containers:
    ax.bar_label(i)
plt.show()

fig.savefig('clear-water.png', facecolor=fig.get_facecolor(), edgecolor='none')
```

```python
# 1 red drop
labels = data1r.columns.to_list()
labels = labels[1:]
bars = []
col = ['purple','blue','cyan','lime','lawngreen','gold','darkorange','red','gray','crimson']
x = np.arange(len(labels))

j = 0
for i in labels:
    bars.append(np.mean(data1r[i].to_numpy())-clear_bars[j])
    j += 1

fig, ax = plt.subplots(figsize=(10,6))

ax.bar(labels, bars, align='center', color=col)
ax.set_xticks(x, labels, rotation=90)
plt.ylabel('Intensity Diference From Clear Water')
plt.title('1 red drop (100ml)')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()

fig.savefig('1-drop-r.png', facecolor=fig.get_facecolor(), edgecolor='none')
```
